```python
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
plt.style.use('ggplot')
```

```python
data = pd.read_csv("cities.csv")
data.head()
```

```python
# Latitude vs: Max temp, wind speed, humidity, cloudiness
lat = data["Lat"]

fig1, ax1 = plt.subplots(figsize = (5,5))
ax1.scatter(lat, data["Max Temp"],
           facecolors = "blue",
           edgecolors = "black",
           s = 30);
ax1.set(xlabel = "Latitude",
       ylabel = "Max Temperature (F)",
       title = "City Latitude vs Max Temperature")
fig1.savefig("maxtemp.png")
```

```python

fig2, ax2 = plt.subplots(figsize = (5,5))
ax2.scatter(lat, data["Wind Speed"],
           facecolors = "blue",
           edgecolors = "black",
           s = 30);
ax2.set(xlabel = "Latitude",
       ylabel = "Wind Speed (mph)",
       title = "City Latitude vs Wind Speed")
fig2.savefig("windspeed.png")
```

```python

fig3, ax3 = plt.subplots(figsize = (5,5))
ax3.scatter(lat, data["Cloudiness"],
           facecolors = "blue",
           edgecolors = "black",
           s = 30);
ax3.set(xlabel = "Latitude",
       ylabel = "Cloudiness (%)",
       title = "City Latitude vs Cloudiness")
fig3.savefig("cloudiness.png")
```

```python

fig4, ax4 = plt.subplots(figsize = (5,5))
ax4.scatter(lat, data["Humidity"],
           facecolors = "blue",
           edgecolors = "black",
           s = 30);
ax4.set(xlabel = "Latitude",
       ylabel = "Humidity (%)",
       title = "City Latitude vs Humidity")
fig4.savefig("humidity.png")
```
